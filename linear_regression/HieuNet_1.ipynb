```python
# Khai báo các thư viện cần thiết
import numpy as np
import pandas as pd
```

```python
# Khai báo thư viện model
from sklearn.linear_model import LinearRegression
```

```python
# Load data
sales_df = pd.read_csv('advertising_and_sales_clean.csv')
sales_df # quan sát dữ liệu
```

```python
# Bỏ dữ liệu dạng chữ (hoặc chuyển nó về số)
sales_df = sales_df.drop(['influencer'], axis = 1)
```

```python
sales_df
```

```python
# Tachs dữ liệu ra tập X,y để train, test
X_train = sales_df.drop(['sales'], axis = 1).values
y = sales_df['sales'].values
X_train
```

```python
y
```

```python
# Chuẩn hoá dữ liệu
from sklearn.preprocessing import MinMaxScaler, StandardScaler
minmax = MinMaxScaler()
X_train = minmax.fit_transform(X_train)
```

```python
X_train
```

```python
# Chia hai tập train, test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train, y, test_size=0.2)  # test_size = 0.2 nghĩa là tập test chiếm 20%
# cái này nghĩa là từ tập X,y ban đầu thu thập được, chia 80% train và 20% còn lại để kiểm tra độ chính xác mô hình
```

```python
# Training
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# Đánh giá mô hìnH
from sklearn.metrics import mean_squared_error # Sử dụng hàm mất mát này để đánh giá
mean_squared_error(model.predict(X_test), y_test)**0.5 # model.predict là để đoán, mũ 0.5 là để lấy căn

# ý nghĩa của 2939 là kết quả dự đoán sẽ lệch so với thực tế trong khoảng này
```

# II - Logistic và Softmax

Logistic đọc kĩ phần Deeplearning cơ bản mình mới gửi ở chương II, Logistic là trường hợp đặc biệt của softmax khi nhãn có 2, nghĩa là phân loại nhị phân

https://www.youtube.com/watch?v=yIYKR4sgzI8

https://machinelearningcoban.com/2017/02/17/softmax/

```python
# Khai báo thư viện
from sklearn.linear_model import LogisticRegression
```

```python
# Load dữ liệu
import sklearn.datasets
digits = sklearn.datasets.load_digits()
X,y = digits.data, digits.target
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=123)
# Trên đây là một bộ dữ liệu có sẵn trong thư viện cung cấp cho chúng ta, dữ liệu đầu vào là các ảnh đen trắng từ 0 đến 9 và output là dự đoán nhãn của nó
```

```python
# Quan sát dữ liệu
import matplotlib.pyplot as plt

data = pd.DataFrame(X)
data['img'] = y



a = (data[data['img'] == 9])
for i in a.drop(['img'], axis = 1).values:
    plt.imshow(i.reshape(8,8), cmap = 'gray', interpolation='nearest')
    plt.show()
# Các ảnh số 9
```

```python
# Tiến hành training
reg = LogisticRegression(penalty='l1', solver='saga', multi_class='multinomial')
# penalty = 'l1' là dùng L1 cho hàm loss như nói ở trên,   multi_class = 'multinomial' là bài toán này phân loại nhiều nhãn
reg.fit(X_train, y_train)
reg.score(X_test, y_test) # Hàm tính accuaracy trên tập test
```

# III - SVM

```python
# Cũng với bài toán trên nếu dùng SVM thì sẽ như thế này 
from sklearn.svm import SVC, LinearSVC

svc = SVC(probability=True)
svc.fit(X_train, y_train)
print('Train: {}'.format(svc.score(X_train, y_train)))
print('Predict: {}'.format(svc.score(X_test, y_test)))
svc.predict_proba(X)[0,0] # Xác suất
```

```python
# Load thư viện
from sklearn.cluster import KMeans

# Load dữ liệu

iris = (pd.read_csv('Iris.csv')).drop(['Id'], axis = 1)
iris_new = iris.drop(['Species'], axis = 1)
iris # dữ liệu về phân loại loài hoa
```

```python
iris_new # Giả sử mình bỏ nhãn của nó đi ( học không giám sát ) để cho máy nó phân loại
```

```python
X = iris_new.values # lấy dữ liệu
```

```python
# Training
model = KMeans(n_clusters=3) # Khai báo số nhóm mình muốn phân hoạch
model.fit(X)
X_pre = model.predict(X)
```
