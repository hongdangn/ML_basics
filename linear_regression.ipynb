```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Problem: Cho dữ liệu gồm số kilomet đi trên taxi, nhãn tương ứng là giá tiền

```python
#1 Chuẩn bị dữ liệu
X = np.array([
    [2],
    [7],
    [9],
    [3],
    [10],
    [6],
    [1]
    ,[8]
])

ones = np.ones_like(X)

X = np.concatenate([X,ones], axis = 1)
# -------------------------------------------
y = np.array([12,35,41,19,45,28,10,55]).reshape(8,1)
# -------------------------------------------

#2 Chọn ngẫu nhiên theta
theta = np.random.normal(size=2).reshape(2,1)

#3 In ra
print(X)
print("--------------")
print(y)
print("--------------")
print(theta)



```

```python
# 4 Thiết Lập Hàm giả thiết
def hypo(X, theta):
  return X@theta

# np.array([1,2,3])**2
```

```python
hypo(X, theta)
```

Thiết lập hàm mất mát: MSE A.K.A Mean Squared Error

```python
#4 Thiết lập hàm mất mát
def cost(X,y,theta):
  lost = (hypo(X,theta) - y)**2
  lost = np.mean(lost)
  return lost/2
cost(X,y,theta)
```

```python
#5: Thiết lập gradient
def gradient(X,y,theta,m):
  grad = (X.T@(hypo(X,theta) - y))/m
  return grad
```

```python
#6: Cập nhập hyperparamater
m = 8 # kích thước dữ liệu
epoch = 150 # Số lần tiến hành
learning_rate = 0.001 #tốc độ học
```

```python
# Cập nhập theta
cst = []
for i in range(epoch):
  grad = gradient(X,y,theta,m)
  theta = theta - learning_rate * grad
  print("Cost is {} after {} epoch".format(cost(X,y,theta), i+1))
  cst.append(cost(X,y,theta))
```

```python
# Hình ảnh minh hoạ
plt.plot(range(epoch), cst)
plt.xlabel("epoch")
plt.ylabel("cost");
```

```python
# Dự đoán giá trị
predict = np.array([40,1])
hypo(predict, theta)
```
